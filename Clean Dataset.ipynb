## Converting Raw Data into a Numerical Usable Format For Analysis

### - Dataset obtained 15/06/2023 from https://www.kaggle.com/datasets/itachi9604/disease-symptom-description-dataset, last updated in 2020 by Pranay Patil

```python
# import necessary libraries
import numpy as np
import pandas as pd
```

```python
# read in raw dataset
data = pd.read_csv("dataset.csv")
```

```python
data.columns[1:]
```

```python
#Obtain all unique labels for symptoms
symp = pd.read_csv("Symptom-severity.csv")
```

```python
rel_symp = symp.columns[0]
```

```python
symptoms = []
for i in range(len(symp[rel_symp])):
    symptoms.append(symp[rel_symp][i])
```

```python
symptoms
```

```python
data.shape
```

```python
# Replace null values with "None"
data.fillna("None", inplace = True)
```

```python
symp_list = {}
ndv = 1000 #Only do 1000 rows for processing to be efficient
#Across every row, if a symptom appears than under a newly created column called by symptom name, place 1 else place 0.
for j in symptoms:
    symp_list[j]=[]
    for l in range(ndv):
        count = 0
        for k in range(data.shape[1]-1):
            if data[f"Symptom_{k+1}"][l].strip().lower() == j.strip().lower():
                count += 1
        symp_list[j].append(count)
```

```python
symp_list
```

```python
#Convert symptoms and their associated values into a dataframe
final_data_frame = pd.DataFrame.from_dict(symp_list)
```

```python
final_data_frame.head()
```

```python
#Create new column with the associated diseases for each row
final_data_frame["Disease"] = data["Disease"][:ndv]
```

```python
final_data_frame.head()
```

```python
#Convert data to a csv and save it
final_data_frame.to_csv('clean_symptoms.csv', index=False)
```
